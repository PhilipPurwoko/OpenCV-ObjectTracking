# Prepare Data

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/pen.jpeg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

# Process Image

```python
# Convert image to gray
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
# Gamma corection
gray_correct = np.array(255 * (gray / 255) ** 1.2 , dtype='uint8')
```

```python
plt.imshow(gray_correct,cmap='gray')
```

```python
# Apply threshold
thresh = cv2.adaptiveThreshold(gray_correct, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 255, 19)
thresh = cv2.bitwise_not(thresh)
```

```python
plt.imshow(thresh,cmap='gray')
```

```python
# Dilatation + erosion
kernel = np.ones((15,15), np.uint8)
image_dilatation = cv2.dilate(thresh, kernel, iterations=1)
image_erosion = cv2.erode(image_dilatation,kernel, iterations=1)

# clean all noise after dilatation and erosion
iamge_erosion = cv2.medianBlur(image_erosion, 7)
```

```python
plt.imshow(image_erosion,'gray')
```

# Labeling

```python
ret, labels = cv2.connectedComponents(image_erosion)
object_count = str(ret-1)

label_hue = np.uint8(179 * labels / np.max(labels))
blank_ch = 255 * np.ones_like(label_hue)

labeled_img = cv2.merge([label_hue, blank_ch, blank_ch])
labeled_img = cv2.cvtColor(labeled_img, cv2.COLOR_HSV2BGR)
labeled_img[label_hue == 0] = 0
```

```python
plt.imshow(labeled_img)
plt.title(f'Object counts : {object_count}')
```
